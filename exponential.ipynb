```python
import pandas as pd
import numpy as np
import seaborn as sns
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv('exponential.csv')
XP = raw.iloc[:,[True, False]]
curve = raw.iloc[:, [False, True]]
XParray = XP.to_numpy()
curvearray = curve.to_numpy()
xdata = XParray.flatten()
ydata = curvearray.flatten()
```

```python
a = sns.scatterplot(x=xdata, y=ydata)
```

```python
def func(x, a, b):
    return a * (1-(np.exp(-b * x)))
```

```python
popt, pcov = curve_fit(func, xdata, ydata, bounds=(0, [1000., 1.]))
```

```python
b = sns.scatterplot(x=xdata, y=ydata)
b = sns.lineplot(x=xdata, y=func(xdata,*popt), color='red')
```
